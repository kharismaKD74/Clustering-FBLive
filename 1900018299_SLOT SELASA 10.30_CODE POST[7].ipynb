```python
#import library yang digunakan
import pandas as pd
from sklearn.cluster import KMeans  #metode clustering
import matplotlib.pyplot as plt  #untuk grafik tampil data
import numpy as np
```

```python
#import dataset
FB = pd.read_csv('FBLive.csv')
```

```python
FB
```

```python
#mengecek atribut yang ada NA dan tidak
FB.isna().sum()
```

# Seleksi data

```python
#mengambil data di atribut num_reactions dan num_comments 
X = FB[['num_reactions','num_comments']]
X
```

```python
#mengambil data index ke 0 dan 1 pada X
plt.plot(X.iloc[:,0], X.iloc[:,1], 'go')
```

# Clustering

```python
#clustering dengan k=5
kMeans5 = KMeans(n_clusters=5)
kMeans5.fit(X) #fitting data
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=kMeans5.labels_, marker='o') #c=kelas, marker=membentuk data pada grafik
plt.show()
```

```python
#membuat kolom baru untuk menampung hasil kluster5
print(kMeans5.labels_)

X["kluster5"] = kMeans5.labels_
X.head()
```

```python
#clustering dengan k=10
kMeans10 = KMeans(n_clusters=10)
kMeans10.fit(X)  #fitting data
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=kMeans10.labels_, marker='o') #c=kelas, marker=membentuk data pada grafik
plt.show()
```

```python
#membuat kolom baru untuk menampung hasil kluster10
print(kMeans10.labels_)

X["kluster10"] = kMeans10.labels_
X.head()
```

```python
#clustering dengan k=15
kMeans15 = KMeans(n_clusters=15)
kMeans15.fit(X)  #fitting data
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=kMeans15.labels_, marker='o') #c=kelas, marker=membentuk data pada grafik
plt.show()
```

```python
#membuat kolom baru untuk menampung hasil kluster15
print(kMeans15.labels_)

X["kluster15"] = kMeans15.labels_
X.head()
```
